```python
import csv

counter = 0

with open('Airbnb_Texas_Rentals.csv', 'r',encoding="utf8") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row: # if row is not empty, write a file with this row
            filename = "doc_%s.tsv" % str(counter)
            with open(filename, 'wb',encoding="utf8") as csvfile_out:
                writer = csv.writer(csvfile_out,delimiter="\t")
                writer.writerow(row[1:])
                counter += 1
```

```python
#TEXT CLEANING  -- only description. making a function?
#appending title to description . fix \n

words = []

for i in range (1,4): #must be corrected

    myfile = open('/Users/mattiasbasso/Desktop/HW3/docs/doc_'+str(i)+'.tsv', 'r')
    txt = myfile.read().split('\t')
    description = txt[4]
    #description = description.replace('\n', ' ')
    
    #punctuation
    description = description.translate(str.maketrans(' ', ' ', string.punctuation))
    title = txt[7]
    print(description)
   
    #tokenizing
    tokens_d = nltk.word_tokenize(description)
    print('TOKEN')
    print(tokens_d)
    
    
    #stopwords
    stop = set(stopwords.words('english'))
    tokens_d = [w for w in tokens_d if not w in stop]
    print('STOPWORDS')
    print(tokens_d)
    
    
    #lower case ?????
    tokens_d = [w.lower() for w in tokens_d]
    print('LOWERCASE')
    print(tokens_d)
    
    #stemming - look deeply
    #ps = PorterStemmer()
    #tokens_d = [ps.stem(w) for w in tokens_d]
    print('STEMMING')
    print(tokens_d)
    
    #append all words we find in a list
    for word in tokens_d:
        words.append(word)
    
    #print('/')
```

```python
#deleting duplicates
words = set(words)
```

```python
#creating vocabulary

vocabulary = {}
count = 1
for word in words:
    #print(word)
    vocabulary[word] = count
    count += 1
```
